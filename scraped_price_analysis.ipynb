```python
price_data_multi_order['Review_Count_Avg'] = price_data.groupby('Name')['Review Count'].mean()
price_data_multi_order['Score_Avg'] = price_data.groupby('Name')['User Review Score'].mean()
price_data_multi_order['Price_Avg'] = price_data.groupby('Name')['Price'].mean()
price_data_multi_order.head(20)
```

```python
price_data_multi_order['Review_Count_Difference'] = price_data_multi_order['Review Count'] - price_data_multi_order['Review_Count_Avg']
price_data_multi_order['Score_Difference'] = price_data_multi_order['User Review Score'] - price_data_multi_order['Score_Avg']
price_data_multi_order['Price_Difference'] = price_data_multi_order['Price'] - price_data_multi_order['Price_Avg']
price_data_multi_order.head(30)
# Lower the better
```

```python
price_data_multi_order['Review_Count_Difference_Perc'] = (price_data_multi_order['Review Count'] / price_data_multi_order['Review_Count_Avg']) * 100
price_data_multi_order['Score_Difference_Perc'] = (price_data_multi_order['User Review Score'] / price_data_multi_order['Score_Avg']) * 100
price_data_multi_order['Price_Difference_Perc'] = (price_data_multi_order['Price'] / price_data_multi_order['Price_Avg']) * 100
price_data_multi_order.head(30)
# Lower the better for price, higher the better for score and review count
```

```python
price_data_multi_order = price_data_multi_order.reset_index()
```
